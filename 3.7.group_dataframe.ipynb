## Objective
- learn some basic methods to group DataFrame

```python
# Load the required packages
import pandas as pd

# Read the data
df = pd.read_csv('./data/gdp_china_renamed.csv')

# diplay the first 5 rows
df.head()
```

## 1. grouping by one column and calculate summary statistics of all columns for each group

```python
df.groupby(['prov']).mean().round(3)
```

## 2. grouping by one column and calculate summary statistics of one column for each group

```python
df.groupby(['prov'])[['pop']].mean()
```

```python
# grouped by year and calculate sum population for each year: 
df.groupby(['year'])[['p0op']].sum()
```

## 3. choose to include NA in group keys or not by setting `dropna` parameter

```python
df.groupby(['pop'],dropna=False).sum() #default dropna=True
```
