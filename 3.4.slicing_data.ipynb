## 4. Slicing rows and columns
#### (1) using `.loc`

```python
# using column names
df.loc[0:6,['prov','year','gdp']]
```

```python
# all rows
df.loc[:,['prov','year','gdp']]
```

```python
# exclude one column
df.loc[:,df.columns!='year']
```

#### (2) using `.iloc`

```python
# continous columns
df.iloc[0:6,0:4]
```

```python
# discontinuous columns
df.iloc[0:6,[0,2,4]]
```

## 5. `loc` vs `iloc`

- `loc`: label-based, which specify rows and columns based on their row and column labels.

- `iloc`: integer position-based, which specify rows and columns by their integer position values (0-based integer position).
