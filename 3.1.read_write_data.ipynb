### (2) Read online data

````python
url ='https://raw.githubusercontent.com/WEIAI-TECH/python-data-analysis-modelling_beginner/main/data/gdp_china_clean.csv'
df2 = pd.read_csv(url)
````

````python
df2.head()
````

````python
df2.tail()
````

There is a number of pandas commands to read and write other data formats, such as:
```python
pd.read_excel('filename.xlsx',sheet_name='Sheet1', index_col=None, na_values=['NA'])
pd.read_stata('filename.dta')
pd.read_sas('filename.sas7bdat')
pd.read_hdf('filename.h5','df')

...
```
Note: The above command has many optional arguments to fine-tune the data import process.

More information can be refers to https://pandas.pydata.org/pandas-docs/stable/user_guide/io.html

### (4) Skip rows from top

````python
df2 = pd.read_csv(url,skiprows = 2)
````

````python
df2.head()
````

### (5) Skip rows from bottom / footer

````python
df2 = pd.read_csv(url,skipfooter =7,engine='python')
````

````python
df2.tail()
````

````python
df2 = pd.read_csv(url,skiprows=2,skipfooter =7,engine='python')
````

````python
df2.tail()
````

## 3 Save DataFrame

````python
df2.to_csv('./data/gdp_china_clean.csv', index=False)
````
