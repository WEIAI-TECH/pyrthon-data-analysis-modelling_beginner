## Objective
- learn some basic methods to sort Panadas DataFrame

```python
# Load the required packages
import pandas as pd

# Read the data
df = pd.read_csv('./data/gdp_china_renamed.csv')

# diplay the first 5 rows
df.head()
```

## 1. Sorting the data by one column

```python
# default ascending=True
df.sort_values(by=['gdp'],ascending=False)
```

```python
# or omit 'by='
df.sort_values(['gdp'],ascending=False)
```

## 2. Sorting the data by 2 or more columns

```python
df.sort_values(by=['year','gdp'],ascending=[False,True]) 
```

## 3. Sorting the Data with Putting NAs first

```python
df.info()
```

```python
df.sort_values(by=['pop'],na_position='first')
```
